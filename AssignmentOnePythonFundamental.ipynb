```python
# Problem 15 - same data, 3 formatting styles
product = 'Laptop'
price = 75000.5
discount = 10

print(f"Product: {product} | Price: {price:.2f} | Discount: {discount}%")
print("Product: {} | Price: {:.2f} | Discount: {}%".format(product, price, discount))
print("Product: %s | Price: %.2f | Discount: %d%%" % (product, price, discount))
```

```python
# Problem 17 - rectangle area/perimeter calculator
length = float(input("Enter length: "))
width = float(input("Enter width: "))
area = length * width
perimeter = 2 * (length + width)
print(f"Area: {area:.2f}")
print(f"Perimeter: {perimeter:.2f}")
```

```python
# Problem 18 - grade scale + pass/fail
score = int(input("Enter your score (0-100): "))

if score >= 90:
    grade = 'A+'
elif score >= 80:
    grade = 'A'
elif score >= 60:
    grade = 'B'
elif score >= 40:
    grade = 'C'
else:
    grade = 'F'
print(f"Grade: {grade}")

if score >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
# Problem 19 - login check, nested if-else
correct_username = "ash ketchup"
correct_password = "pikachu"

entered_username = input("Username: ")
entered_password = input("Password: ")

if entered_username == correct_username:
    if entered_password == correct_password:
        print("Access Granted")
    else:
        print("Wrong Password")
else:
    print("User Not Found")
```

```python
# Problem 20 - ternary for sign, nested if for even/odd
number = float(input("Enter a number: "))

sign = "Positive" if number > 0 else ("Negative" if number < 0 else "Zero")
print(sign)

if number != 0:
    if number % 2 == 0:
        print("Even")
    else:
        print("Odd")
else:
    print("N/A — zero has no even/odd check")
```
