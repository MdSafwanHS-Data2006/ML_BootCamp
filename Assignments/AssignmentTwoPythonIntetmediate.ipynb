```python
#Q1 - sum all numbers using a loop
numbers = [4, 15, 7, 23, 11]
total = 0
for num in numbers:
    total = total + num
print("Sum:", total)
```

```python
#Q2 - first and last item via index
fruits = ["apple", "mango", "banana", "grape", "orange"]
print("First:", fruits[0])
print("Last:", fruits[-1])
```

```python
#Q3 - count how many numbers are greater than 10
nums = [5, 12, 3, 18, 7, 25, 9]
count = 0
for n in nums:
    if n > 10:
        count = count + 1
print("Count greater than 10:", count)
```

```python
#Q14 - positive/negative, then even/odd if positive
num = int(input("Enter a number: "))
if num > 0:
    print("Positive")
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
else:
    print("Negative")
```

```python
#Q15 - pass/fail, then grade if passed
marks = int(input("Enter marks: "))
if marks >= 40:
    print("Pass")
    if marks >= 80:
        print("Grade: A")
    elif marks >= 60:
        print("Grade: B")
    else:
        print("Grade: C")
else:
    print("Fail")
```

```python
#Q16 - adult check, then senior citizen check
age = int(input("Enter age: "))
if age > 18:
    print("Adult")
    if age > 60:
        print("Senior Citizen")
    else:
        print("Not a Senior Citizen")
else:
    print("Not an Adult")
```

```python
#Q17 - print 1 to n using a loop inside the function
def print_numbers(n):
    for i in range(1, n + 1):
        print(i)
print_numbers(5)
print_numbers(10)
```

```python
#Q18 - count down from n to 1
def count_down(n):
    for i in range(n, 0, -1):
        print(i)
count_down(5)
count_down(3)
```

```python
#Q19 - sum a list using a loop inside the function
def sum_of_list(numbers):
    total = 0
    for num in numbers:
        total = total + num
    return total
print(sum_of_list([1, 2, 3, 4, 5]))
print(sum_of_list([10, 20, 30]))
```

```python
#Q20 - multiplication table from 1 to 10
def print_table(number):
    for i in range(1, 11):
        print(number, "x", i, "=", number * i)
print_table(5)
print_table(7)
```
